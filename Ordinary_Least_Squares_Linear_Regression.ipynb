```python
import pandas as pd
import numpy as np
```

```python
st_scores = pd.read_csv('student_scores.csv')
st_scores
```

```python
a = st_scores['Hours']
b = st_scores['Scores']

```

```python
#Ordinary least squares
def ols_func(a,b):
    a_mean = a.mean()
    b_mean = b.mean()
    x1 = np.sum((b-b_mean)*(a-a_mean))/np.sum((a-a_mean)**2) 
    x0 = b_mean- x1*a_mean
    return(x0,x1)

def predict(x,y):
    predictions = list()
    for hours in a:
        y_hat = x0 + x1 * hours 
        predictions.append(y_hat) 
    return predictions
    
    
```

```python
x0,x1 = ols_func(a,b)
print('Slope: %f Intercept: %f' %(x1,x0))
```
